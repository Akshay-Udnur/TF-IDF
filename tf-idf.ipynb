### TFIDF

```python
import numpy as np
class TF_IDF():
    def __init__(self,sentences):
        # First remove punctuations and also make all the words in small letters.
        sentences = [i.replace(",","") for i in sentences]
        sentences = [i.replace(".","") for i in sentences]
        sentences = [i.lower() for i in sentences]
        self.sentences = sentences
        
    def TF(self):
        out = {}
        for sentence in self.sentences:
            tf = {}
            for i in sentence.split():
                tf[i] = sentence.count(i)/len(sentence.split())
            out[sentence]= tf
        return out
    
    def IDF(self):
        out = {}
        for sentence in self.sentences:
            idf = {}
            for i in sentence.split():
                count = 0
                for j in self.sentences:
                    if i in j:
                        count+=1
                idf[i] = np.log(len(self.sentences)/count)
            out[sentence] = idf
        return out
    
    def TFIDF(self):
        out = {}
        tf = self.TF()
        idf = self.IDF()
        for sentence in self.sentences:
            tf_sent = tf[sentence]
            idf_sent = idf[sentence]
            tfidf = {}
            for i in tf_sent:
                tfidf[i]=tf_sent[i]*idf_sent[i]
            out[sentence] = tfidf
        return out
```

```python
sentences = ["The iPhone battery life is exceptional, easily lasting throughout the day even with heavy usage.",
             "The camera quality on the iPhone is outstanding, capturing stunning photos and videos in various lighting conditions.",
             "The A15 Bionic chip in iPhone delivers remarkable performance, ensuring smooth multitasking and fast app responsiveness.",
             "The design of the iPhone feels sleek and premium, offering an excellent user experience.",
             "The new features like Cinematic mode add a creative touch to the iPhone camera, enhancing its capabilities.",
             "I am impressed with the display quality of the iPhone, providing crisp and vibrant visuals."]

temp = TF_IDF(sentences)
```
